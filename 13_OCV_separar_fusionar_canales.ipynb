# División y Fusión de Canales con OpenCV

### Importar paquetes

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### Function to display images in Jupyter Notebooks and Google Colab

```python
def mostrar_imagen(title, imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

### How to split and merge channels with OpenCV


```python
args = {
	"imagen": "wave.png"
}
```

```python
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original", imagen)
```

```python
(B, G, R) = cv2.split(imagen)

# mostrar cada canal individualmente
mostrar_imagen("Red", R)
mostrar_imagen("Green", G)
mostrar_imagen("Blue", B)
```

```python
fusionado = cv2.merge([B, G, R])
mostrar_imagen("Fusionado", fusionado)
```

```python
# visualizar cada canal a color
zeros = np.zeros(imagen.shape[:2], dtype="uint8")
mostrar_imagen("Red", cv2.merge([zeros, zeros, R]))
mostrar_imagen("Green", cv2.merge([zeros, G, zeros]))
mostrar_imagen("Blue", cv2.merge([B, zeros, zeros]))
```
