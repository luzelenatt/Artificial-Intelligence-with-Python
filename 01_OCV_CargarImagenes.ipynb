```python
from matplotlib import pyplot as plt
import cv2
```

Cargar im치genes y mostrarlas en Jupyter Notebook

```python
#carga la imagen y muestra informaci칩n sobre esta
imagen = cv2.imread("golden_retriever.jpg")
(alto,ancho,canales) = imagen.shape
print("alto: {} pixeles".format(imagen.shape[0]))
print("ancho: {} pixeles".format(imagen.shape[1]))
print("canales: {}".format(imagen.shape[2]))
```

```python
#convierte la imagen de un formato de color BGR a RGB y lo mostramos
imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
plt.title("Golden")
plt.grid(False)
plt.show
```

```python
#Guardar la imagen en formato RGB
cv2.imwrite("nuevaImagen02.jpg",cv2.cvtColor(imagen,cv2.COLOR_RGB2BGR))
```
