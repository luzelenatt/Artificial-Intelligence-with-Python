Importar paquetes y librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

Función para mostrar una imagen con imshow() de pyplot

```python
def mostrar_imagen(title, imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

Aritmética de imágenes con OpenCV

```python
args = {
    "imagen" :"labrador.jpg"
}
```

Las imágenes son matrices NumPy almacenadas como enteros de 8 bits sin signo (uint8) con valores en el rango [0, 255]; al usar las funciones de suma / resta en OpenCV, estos valores se recortarán a este rango, incluso si caen fuera del rango [0, 255] después de aplicar la operación

```python
suma = cv2.add(np.uint8([200]), np.uint8([100]))
resta = cv2.subtract(np.uint8([50]), np.uint8([100]))
print("suma: {}".format(suma))
print("resta: {}".format(resta))
```

El uso de operaciones aritméticas NumPy (en lugar de operaciones OpenCV) dará como resultado el módulo(resto)
 en lugar de ser recortado al rango [0, 255]

```python
suma = np.uint8([200]) + np.uint8([57])
resta = np.uint8([50]) - np.uint8([51])
print("resultado suma: {}".format(suma))
print("resultado resta: {}".format(resta))
```

Cargar la imagen de entrada original y mostrarla en nuestra pantalla

```python
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

El aumento de las intensidades de píxeles en nuestra imagen de entrada en 100 se logra construyendo una matriz NumPy que tiene las mismas dimensiones que nuestra imagen de entrada, llenándola con unos, multiplicándola por 100 y luego sumando la imagen de entrada y la matriz juntas

```python
imagen.shape
```

```python
np.ones((3,2),dtype='uint8')*100
```

```python
M = np.ones(imagen.shape, dtype="uint8") * 100
sumado = cv2.add(imagen, M)
mostrar_imagen("Más claro", sumado)
```

Del mismo modo, podemos restar 50 de todos los píxeles de nuestra imagen y hacerla más oscura.

```python
M = np.ones(imagen.shape, dtype="uint8") * 50
restado = cv2.subtract(imagen, M)
mostrar_imagen("Más Oscuro", restado)
```

```python
M = np.ones(imagen.shape, dtype="uint8") * 100
sumado = imagen + M
mostrar_imagen("Módulo", sumado)
```

```python
M = np.ones(imagen.shape, dtype="uint8") * 100
restado = imagen - M
mostrar_imagen("Módulo", restado)
```
