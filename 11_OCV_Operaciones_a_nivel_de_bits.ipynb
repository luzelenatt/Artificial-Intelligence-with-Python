# Operaciones a nivel de Bits: AND, OR, XOR y NOT

Importar paquetes y librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

Función para mostrar una imagen con imshow() de pyplot

```python
def mostrar_imagen(title, imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

Operadores AND, OR, XOR y NOT en OpenCV

```python
# lienzo
lienzo_rect = np.zeros((300,300),dtype='uint8')
mostrar_imagen("Lienzo", lienzo_rect)
```

```python
#dibujar rectángulo
cv2.rectangle(lienzo_rect,(25,25),(275,275),255,-1)
mostrar_imagen("Rectángulo", lienzo_rect)
```

```python
#dibujar circulo
lienzo_circ = np.zeros((300,300),dtype='uint8')
cv2.circle(lienzo_circ,(150,150),150,255,-1)
mostrar_imagen("Circulo", lienzo_circ)
```

El resultado de la operación AND a nivel de bits es "True" si ambos valores tienen un valor distinto de cero, por lo tanto el pixel se activa (ej: toma un valor de 255). En caso contrario el pixel se desactiva (ej: toma un valor de 0)

```python
andImagenes = cv2.bitwise_and(lienzo_rect,lienzo_circ)
mostrar_imagen("Operador AND", andImagenes)
```

El resultado de la operación OR a nivel de bits es "True" si alguno de los valores tienen un valor distinto de cero, por lo tanto el pixel se activa (ej: toma un valor de 255). En caso contrario el pixel se desactiva (ej: toma un valor de 0)

```python
orImagenes = cv2.bitwise_or(lienzo_rect,lienzo_circ)
mostrar_imagen("Operador OR", orImagenes)
```

El resultado de la operación XOR a nivel de bits es "True" si alguno de los valores tienen un valor distinto de cero y el otro cero, por lo tanto el pixel se activa (ej: toma un valor de 255). En caso contrario el pixel se desactiva (ej: toma un valor de 0)

```python
xorImagenes = cv2.bitwise_xor(lienzo_rect,lienzo_circ)
mostrar_imagen("Operador XOR", xorImagenes)
```

NOT invierte el valor del pixel. Si es 255 se convierte a cero y viceversa

```python
notCirculo = cv2.bitwise_not(lienzo_circ)
mostrar_imagen("NotCirculo",notCirculo)
```
